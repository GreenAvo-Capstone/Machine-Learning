## Dataset Description 


This dataset consists of a collection of avocado images categorized into three main categories: “RAW” for images showing avocados in their raw condition, "half-cooked" for images showing avocados in a half-ripe condition, and “RIPE” for images showing avocados in a ripe condition. Each image is organized into a folder corresponding to its label. This data will be used to train and test a Convolutional Neural Network (CNN) model using the ResNet-50 architecture. The primary goal is to automatically classify avocado images.
